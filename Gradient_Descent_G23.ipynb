Initial parameters and Data Points

```python
import numpy as np
import matplotlib.pyplot as plt


X = np.array([1, 3])
Y = np.array([3, 6])

m = -1.0
b = 1.0

n = len(X) # No of data points
learning_rate = 0.1

num_iterations = 4

print(f"Initial m: {m}, Initial b: {b}\n")
```

Lists to store the values of *m₀ and b₀* and *MSE* at each iteration

```python
m_history = []
b_history = []
mse_history = []
```

Iteration Loop for Gradient Descent

*y = mx + b*

```python
for i in range(num_iterations + 1): # +1 to include initial state in history

    # Store current parameters and compute MSE before update
    m_history.append(m)
    b_history.append(b)

    # Compute predictions (ŷ) using current m and b
    Y_pred = m * X + b

    # Compute Mean Squared Error (MSE)
    mse = np.mean((Y - Y_pred)**2)
    mse_history.append(mse)

    print(f"--- Iteration {i} ---")
    print(f"Current m: {m:.4f}, Current b: {b:.4f}")
    print(f"Predictions (ŷ): {Y_pred}")
    print(f"Current MSE: {mse:.6f}")

    # Only update if more iterations are planned
    if i < num_iterations:
        errors = Y - Y_pred
        grad_m = -(2/n) * np.sum(errors * X)
        grad_b = -(2/n) * np.sum(errors)

        # Update m and b
        m_new = m - learning_rate * grad_m
        b_new = b - learning_rate * grad_b

        m = m_new
        b = b_new

        print(f"Gradients: ∂J/∂m = {grad_m:.4f}, ∂J/∂b = {grad_b:.4f}")
        print(f"Updated m: {m:.4f}, Updated b: {b:.4f}\n")
    else:
        print("\nGradient Descent complete.\n")
```

Compute predicitions using final parameters

```python
print("--- Final Predictions ---")
final_m = m_history[-1]
final_b = b_history[-1]
final_Y_pred = final_m * X + final_b
print(f"Final m: {final_m:.4f}, Final b: {final_b:.4f}")
print(f"Final predictions (ŷ): {final_Y_pred}")
print(f"Final MSE: {mse_history[-1]:.6f}")
```

Visualization

```python
iterations = range(num_iterations + 1)

# Plot 1: m and b change over iterations
plt.figure(figsize=(10, 6))
plt.plot(iterations, m_history, marker='o', linestyle='-', color='blue', label='m (Slope)')
plt.plot(iterations, b_history, marker='x', linestyle='--', color='red', label='b (Intercept)')
plt.title('Change in m and b over Iterations')
plt.xlabel('Iteration Number')
plt.ylabel('Parameter Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Error (MSE) change over iterations
plt.figure(figsize=(10, 6))
plt.plot(iterations, mse_history, marker='o', linestyle='-', color='green', label='MSE (Mean Squared Error)')
plt.title('Change in Mean Squared Error (MSE) over Iterations')
plt.xlabel('Iteration Number')
plt.ylabel('MSE Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
